```python
import cv2
from win32api import GetSystemMetrics
import numpy as np
image = cv2.imread("image/2.jpg", cv2.IMREAD_COLOR)
# 이미지 크기
(h, w) = image.shape[:2]

# 위도우 크기
x = GetSystemMetrics(0) / w * 0.8
y = GetSystemMetrics(1) / h * 0.8

if x<y:
    image = cv2.resize(image, (0,0), fx=x, fy=x)
else:
    image = cv2.resize(image, (0,0), fx=y, fy=y)

# 색 범위
lower_blue = (100, 20, 20)
upper_blue = (130, 255, 255)

dst = image.copy()# 원본 카피
hsv = cv2.cvtColor(dst, cv2.COLOR_BGR2HSV) # hsv로 변환
gray = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY) # gray로 변환
mask = cv2.inRange(hsv, lower_blue, upper_blue) # 지정 범위로 색 찾기

# 최대크기 컨투어 찾기
contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
c = max(contours, key=cv2.contourArea)

mask = np.zeros_like(dst) # mask 비우기

cv2.drawContours(mask, [c], 0, (255, 255, 255), -1) # 빈 mask에 채운 컨투어 그리기(-1)

gray2 = cv2.merge((gray, gray, gray)) # gray 3채널로 확장

gray2 = cv2.bitwise_and(gray2, mask) # mask 모양만 gray2에서 뽑기

# gray2를 기준으로 dst와 필셀 단위로 비교, 값이 있으면 gray2의 픽셀 출력, 없으면 dst의 픽셀 출력
dst = np.where(gray2 != 0, gray2, dst)

cv2.imshow("image", image)
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("image/2.jpg", cv2.IMREAD_COLOR)
# 이미지 크기
(h, w) = image.shape[:2]

# 위도우 크기
x = GetSystemMetrics(0) / w * 0.8
y = GetSystemMetrics(1) / h * 0.8

if x<y:
    image = cv2.resize(image, (0,0), fx=x, fy=x)
else:
    image = cv2.resize(image, (0,0), fx=y, fy=y)

# 색 범위
lower_blue = (100, 20, 20)
upper_blue = (130, 255, 255)

dst = image.copy()# 원본 카피
hsv = cv2.cvtColor(dst, cv2.COLOR_BGR2HSV) # hsv로 변환
gray = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY) # gray로 변환
mask = cv2.inRange(hsv, lower_blue, upper_blue) # 지정 범위로 색 찾기

# 최대크기 컨투어 찾기
contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
c = max(contours, key=cv2.contourArea)

mask = np.zeros_like(dst) # mask 비우기

cv2.drawContours(mask, [c], 0, (255, 255, 255), -1) # 빈 mask에 채운 컨투어 그리기(-1)

# gray 3채널로 확장
gray2 = np.zeros_like(dst)
gray2[:,:,0] = gray
gray2[:,:,1] = gray
gray2[:,:,2] = gray

gray2 = cv2.bitwise_and(gray2, mask) # mask 모양만 gray2에서 뽑기
mask_inv = cv2.bitwise_not(mask) # mask 반전
dst = cv2.bitwise_and(dst, mask_inv) # 배경만 뽑기
dst = cv2.bitwise_or(dst, gray2) # 원본에 합치기

cv2.imshow("image", image)
cv2.imshow("dst", dst)
cv2.imshow("hsv", hsv)
cv2.imshow("gray2", gray2)
cv2.imshow("mask", mask)
cv2.imshow("mask_inv", mask_inv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np    
from win32api import GetSystemMetrics

# opencv 한글 경로 지원 하지 않음
# numpy를 통해 우회하는 경로로 파일 로드
def imread(filename, flags=cv2.IMREAD_COLOR, dtype=np.uint8): 
    try: 
        n = np.fromfile(filename, dtype) 
        img = cv2.imdecode(n, flags) 
        return img 
    except Exception as e: 
        print(e) 
        return None

def FitToWindowSize(image):
    image_resized = image.copy()
    #윈도우 크기 얻기
    # from win32api import GetSystemMetrics
    # print("Width =", GetSystemMetrics(0))
    # print("Height =", GetSystemMetrics(1))
    #이미지 크기 얻기
    print('image {}'.format(image.shape))
    win_w=GetSystemMetrics(0)
    win_h=GetSystemMetrics(1)
    img_h, img_w = image.shape[:2]

    if(img_h > win_h or img_w > win_w):   
        rate_width =  (win_w / img_w)*0.95
        rate_height =  (win_h / img_h)*0.95
        scale = rate_width if (rate_width < rate_height) else rate_height
        image_resized = cv2.resize(image, dsize=(0, 0), fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)
    # cv2.imshow('image_resize',image_resized)
    return image_resized
    


# cv2.imread가 아님/ 위에 정의한 imread()함수 사용
filename = 'image/orange.jpg'
# filename = 'images/sausage.jpg'
#filename = 'E:/OneDrive/Jupyter/opencv/PCB2/양품/양품1_2.bmp'
image = imread(filename,cv2.IMREAD_COLOR)
if image is None:
    print("이미지를 읽을 수 없습니다.")
    exit(1)

#image split -> BGR to HSV
# hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
# h, s, v = cv2.split(hsv)

# mask = cv2.inRange(h, 100,107)
# blue = cv2.bitwise_and(hsv, hsv, mask=mask)
# blue = cv2.cvtColor(blue, cv2.COLOR_HSV2BGR)

# convert color to hsv because it is easy to track colors in this color model

ilowH=98; ilowS=106; ilowV=106
ihighH=110 ; ihighS=255; ihighV=255

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lower_blue = (ilowH, ilowS, ilowV)
upper_blue = (ihighH, ihighS, ihighV)
# Apply the cv2.inrange method to create a mask
mask = cv2.inRange(hsv, lower_blue, upper_blue)
# Apply the mask on the image to extract the original color
image = cv2.bitwise_and(image, image, mask=mask)


# rgb <--> hsv : http://colorizer.org/
resized_image = FitToWindowSize(image)
cv2.imshow("image",resized_image)
cv2.waitKey(0)
cv2.destoryAllWindows()
```
