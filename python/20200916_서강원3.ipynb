# 2-5-1

```python
susung_mean = dalsung[dalsung['구'] == '수성구']
susung_mean = susung_mean.groupby('단지명').mean().round(1)
susung_mean
```

```python
susung_top7 = susung_mean['거래금액(만원)'].sort_values(ascending=False)[:7]
susung_top7
```

```python
susung_top7_values = list(np.ravel(susung_top7.values))
x = np.arange(len(susung_top7_values))  # the label locations
width = 0.6  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 5)) #fig, ax = plt.subplots(TOTAL_X,TOTAL_Y,figsize=(8, 5))
plt.grid(b=True, which='both', axis='both')
rects1 = ax.bar(susung_top7.index, susung_top7_values, width, align='center', alpha=0.5, label='거래금액(만원)')
rects1[susung_top7_values.index(max(susung_top7_values))].set_color('g')

ax.set_title('susung_top7')
ax.set_xticks(x)
ax.set_xticklabels(susung_top7.index)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
fig.tight_layout()

plt.show()
```
