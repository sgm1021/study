```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# 함수 정의
def f2(x, w):
    return (x-w)*x*(x+w)
# x를 정의
x=np.linspace(-3,3,100)

# 차트를 묘사
plt.plot(x,f2(x,3), color='chartreuse', label='$w=3$')
plt.plot(x,f2(x,2), color='black', label='$w=2$')
plt.plot(x,f2(x,1), color='cornflowerblue', label='$w=1$')
plt.legend(loc="upper left") #범례 표시
plt.ylim(-25, 25) #y축 범위 지정
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True) #눈금
plt.show()
```

## 그래프 여러개 보여주기
- subplot
- plt.subplot(n1, n2, n)
- n1 : 전체 그림의 세로 개수
- n2 : 가로 그림의 가로 개수
- n : 현재 그림의 위치(왼쪽 위부터 1, 2, 3, ...)  
    : 0이 아니고 1부터 시작

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,5)) #figure 지정
plt.subplots_adjust(wspace=0.3, hspace=0.3) # 그래프 간 간격지정
for i in range(6):
    plt.subplot(2,3,i+1) #그래프의 위치 지정
    plt.title(i+1)
    plt.plot(x,f2(x,i), 'k')
    plt.ylim(-40,40)
    plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2*x0**2+x1**2
    ans = r*np.exp(-r)
    return ans
xn = 50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0), len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
```

```python
print(x0)
print(x1)
print(y)
```

```python
print(np.round(y,1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(5,5))
plt.gray()
plt.pcolor(y)
plt.colorbar()
plt.show()
```

## surface

```python
from mpl_toolkits.mplot3d import Axes3D
xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(5, 5))
ax = plt.subplot(1,1,1, projection='3d')
ax.plot_surface(xx0,xx1,y, rstride=1, cstride=1, alpha=0.3, color ='blue', edgecolor='black')
ax.set_zticks((0,0.2)) #z의 눈금을 0~0.2fh wpgks
ax.view_init(80, -75) #ax.view_init(인수1, 인수2): 인수1 = 상하외전 각도 , 인수2 = 좌우 회전 각도 양수(시계방향)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

xn = 10
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0), len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

xx0, xx1 = np.meshgrid(x0,x1)
plt.figure(1, figsize=(6,6))
cont = plt.contour(xx0, xx1, y, 10, colors='k')
#등고선의 높이를 5단계로 지시

cont.clabel(fmt='%3.2f', fontsize=8)
 #plt.contour의 반환값을 cont에 저장하고 cont.clabel로 출력

plt.xlabel("$X_0$", fontsize=14)
plt.ylabel("$X_1$", fontsize=14)
plt.show()
```

## vectorize

```python
import numpy as np
matrix = np.arange(1,10).reshape(3,3)
print(matrix)
```

```python
add_100 = lambda i:i+100
```

```python
vectorized_add_100 = np.vectorize(add_100)
vectorized_add_100(matrix)
```

```python
find_odd = lambda i : i if i%2==1 else 0 
find_even = lambda i : i if i%2==0 else 0 
vectorized_find_odd = np.vectorize(find_odd)
vectorized_find_even = np.vectorize(find_even)
print(vectorized_find_odd(matrix))
print(vectorized_find_even(matrix))
```

```python
import numpy as np
matrix= np.arange(1,10).reshape(3,3)

print(matrix)
print(np.max(matrix))
print(np.min(matrix))
print(np.max(matrix, axis=1)) # 각 행에서 최대값 찾음
print(np.min(matrix, axis=1)) # 각 열에서 최소값 찾음
```

```python
vector_column = np.max(matrix, axis=1, keepdims=True)
print(vector_column)
print(vector_column.shape)
```

```python
print(matrix.shape)
print(matrix - vector_column)
```

```python
import numpy as np
matrix= np.arange(1,10).reshape(3,3)
print(matrix)
print(np.mean(matrix))
print(np.mean(matrix, axis=0)) # 각열의 평균
print(np.var(matrix)) # 분산
print(np.std(matrix)) # 표준편차
```

```python
print(np.std(matrix))
print(np.std(matrix, ddof=1))
```

## inverse matrix

```python
matrix = np.array([[1,4],[2,5]])
print(matrix)
```

```python
#역행렬
inv_matrix = np.linalg.inv(matrix)
print(inv_matrix)
```

```python
matrix@inv_matrix
```

```python
matrix = np.arange(1,7).reshape(2,3)
print(matrix)
```

```python
inv_matrix = np.linalg.pinv(matrix)
print(inv_matrix)
```

```python
matrix @ inv_matrix
```

```python
print(np.round(matrix @ inv_matrix,0))
```
